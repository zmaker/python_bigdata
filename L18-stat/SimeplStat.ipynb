## Funzioni statistiche in Python ##

Capire come i dati sono organizzati/disposti. Quali sono le loro caratteristiche essenziali. Cercare degli indicatori che ci facciano capire come sono disposti.

Vediamo come tracciare un semplice grafico:

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()             # Create a figure containing a single Axes.
x = [1,2,3,4,5,6,7]
y = [2,3,4,2,2,3,1]
ax.plot(x,y)  # Plot some data on the Axes.
plt.show() 
```

Usiamo quanto appreso. Carichiamo un file csv con dati sull'etÃ 

```python
import csv
x = []
y = []
with open("persone.csv") as f:
    cur = csv.reader(f, delimiter=',')
    lista = list(cur)  
    i = 0
    for ll in lista:
        if i > 0:
            x.append(i)
            y.append(float(ll[1]))            
        i += 1
```

```python
fig, ax = plt.subplots() 
ax.plot(x,y)  
plt.show() 
```

```python
import statistics

print("media", statistics.mean(y))
print("mediana", statistics.median(y))
print("min", min(y))
print("max", max(y))

print("varianza", statistics.variance(y))
print("varianza_cost", statistics.variance([1,1,1,1,1,1,1,1,1]))
```

```python
fig, ax = plt.subplots() 

media = statistics.mean(y)
# Aggiungi una linea orizzontale per la media
plt.axhline(y=media, color='r', linestyle='--')

mediana = statistics.median(y)
plt.axhline(y=mediana, color='g', linestyle='--')

ax.plot(x,y)  
plt.show() 
```
