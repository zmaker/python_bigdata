## Pandas - Indici


come funzionano gli indici python:


```python
lista = [22,33,44,55,66,77,88]
print(0, lista[0])
print(1, lista[1])
print(2, lista[2])
```

```python
lista.pop(0)
print(lista)
print(0, lista[0])
print(1, lista[1])
print(2, lista[2])
```

In una lista gli indici forniscono la posizione. Non sono delle chiavi! 

Per Pandas Ã¨ diverso:

```python
import pandas as pd
tab = pd.read_csv('persone.csv', delimiter=',')
print(tab)
```

La prima colonna sono le "etichette di identificazione della riga" = una chiave unica associata alla riga e non la sua semplice posizione.
Riordiniamo i dati e vediamo cosa accade:

```python
tab2 = tab.sort_values(by="cognome")
print(tab2)
```

Per accedere ad un elemento con la sua chiave uso loc

```python
print(tab2.loc[0])
```

Uso iloc per accedere alla posizione 

```python
print(tab2.iloc[0])
```

index fornisce la chiave della riga alla posizione indicata. La riga alla posizione 2 ha come chiave:

```python
idx = tab2.index[2]
print(idx)
```

Lo slicing funziona con iloc.

Per estrarre un gruppo di dati con posizione (o indice) compreso tra due valori scrivo:

```python
tab3 = tab2.iloc[2:4]
print(tab3)
```

```python
tab4 = tab2.loc[range(0,4)]
print(tab4)
```

restringo la selezione specificando le colonne

```python
tab4 = tab2.loc[range(0,4)][['cognome', 'data_nascita']]
print(tab4)
```
