Creo un dataframe e lo salvo come excel

```python
import pandas as pd
import numpy as np

# Lista dei mesi
mesi = ['Gennaio', 'Febbraio', 'Marzo', 'Aprile', 'Maggio', 'Giugno', 
        'Luglio', 'Agosto', 'Settembre', 'Ottobre', 'Novembre', 'Dicembre']

# Generiamo valori casuali per ingressi e uscite (tra 1000 e 5000 euro)
ingressi = np.random.randint(1000, 5001, size=12)
uscite = np.random.randint(1000, 5001, size=12)

# Creiamo il DataFrame
cc = pd.DataFrame({
    'mese': mesi,
    'entrate': ingressi,
    'uscite': uscite
})

# Mostriamo il DataFrame
cc
```

Per scrivere u nexcel serve insrtallare con pip: xlsxwriter

```python
wr = pd.ExcelWriter('conto.xlsx', engine='xlsxwriter')
cc.to_excel(wr, index=False, sheet_name='conto')
wr.close()
```

```python
wr = pd.ExcelWriter('conto.xlsx', engine='xlsxwriter')
cc.to_excel(wr, index=False, sheet_name='conto')

# Accesso allo oggetto Workbook per formattazioni aggiuntive
workbook = wr.book
worksheet = wr.sheets['conto']

# Formattazione esempio:
format1 = workbook.add_format({'bold': True, 'font_color': 'blue'})
worksheet.write('D1', 'Dati Personali', format1)

wr.close()
```

## SQL e DB 

installare sqlalchemy

```python
from sqlalchemy import create_engine
eng = create_engine('sqlite:///banca.db')
cc.to_sql('conto', eng, index=False)
```

```python
c = eng.connect()
df = pd.read_sql_table('conto', c)
c.close()

df
```

```python
c = eng.connect()
df = pd.read_sql_query('select * from conto where entrate > 2500', c)
c.close()

df
```
