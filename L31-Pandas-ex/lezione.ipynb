```python
import pandas as pd

df = pd.read_csv("HousingPrices-Amsterdam-August-2021.csv")
df.head()
```

```python
df.info()
```

```python
df = df.rename(columns={'Unnamed: 0':'house_id'})
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
df["Price"].plot(kind='hist', bins=100, edgecolor='black')
plt.show()
```

```python
df.plot(kind="scatter", x='Area', y="Price", alpha = 0.1, edgecolor='black', color='green')
plt.show()
```

```python
df.fillna(0, inplace=True)
```

```python
df['price_per_sqm'] = df['Price']/df['Area']
df.head()
```
