## lotteria

```python
import numpy as np

numeri = np.arange(1,91)
numeri
```

```python
np.random.shuffle(numeri)
numeri
```

```python
estrazioni = np.array([])
```

```python
for i in range(5):
    np.random.shuffle(numeri)
    estratti = numeri[0:5]
    print(estratti)
    estrazioni = np.append(estrazioni, estratti)

estrazioni = estrazioni.reshape(5,5)
estrazioni
```

```python
import pandas as pd

dates = pd.date_range("20240901", periods=5)
dates
```

```python
df = pd.DataFrame(data=estrazioni, index=dates)
df
```

```python
for i in range(len(df)):
    print(df.iloc[i])
```

```python
ritardatari = {}
for i in range(len(df)):
    riga = df.iloc[i]
    for num in riga:
        gg = len(df)-i
        print(num, gg)
        ritardatari[num] = gg

ritardatari
```

```python
df = pd.DataFrame(list(ritardatari.items()) ,columns=["Valore", "Conteggio"])
df
```

```python
df = df.sort_values(by="Conteggio")
df
```
