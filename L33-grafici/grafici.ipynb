**Grafico a torta**

```python
# Valori inventati per i 5 settori
valori = [15, 30, 45, 10, 25]
etichette = ['Settore A', 'Settore B', 'Settore C', 'Settore D', 'Settore E']
colori = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0']  # Colori per ogni settore

# Creare il grafico a torta
#mp.figure(figsize=(8, 6))  # Dimensioni della figura
mp.pie(valori, labels=etichette, colors=colori, autopct='%1.1f%%', startangle=90)

# Equalizza il grafico a torta
mp.axis('equal')  

# Titolo
mp.title('Grafico a Torta con 5 Settori')

# Mostra il grafico
mp.show()
```

**grafico a barre**


```python
import matplotlib.pyplot as plt

# Nomi dei partiti e i loro valori
partiti = ['Partito A', 'Partito B', 'Partito C', 'Partito D']
valori = [1500, 2300, 1800, 2200]  # Valori inventati delle elezioni

# Creare il grafico a barre
#plt.figure(figsize=(10, 6))  # Dimensioni della figura
plt.bar(partiti, valori, color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])

# Aggiungere etichette e titolo
plt.xlabel('Partiti')
plt.ylabel('Numero di Voti')
plt.title('Risultati Elettorali di 4 Partiti')
plt.xticks(rotation=45)  # Ruotare le etichette dell'asse x per migliorare la leggibilit

# Mostrare i valori sopra le barre
for i, v in enumerate(valori):
    print(i,v)
    # Aggiunge i valori sopra le barre
    # barra, altezza del testo, testo
    plt.text(i, v+50, str(v), ha='center', va='bottom')  

# Mostrare il grafico
plt.tight_layout()  # Ottimizza il layout
plt.show()
```

**scatter**


```python
import numpy as np
import matplotlib.pyplot as plt

# Generare dati casuali
x = np.random.rand(50) * 100  # 50 valori casuali per l'asse x
y = np.random.rand(50) * 100  # 50 valori casuali per l'asse y

plt.scatter(x, y)

# Aggiungere etichette e titolo
plt.title('Grafico a Dispersione')
plt.xlabel('Valore X')
plt.ylabel('Valore Y')

# Mostrare la griglia
plt.grid()

# Mostrare il grafico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generare dati casuali
x = np.random.rand(50) * 100  # 50 valori casuali per l'asse x
y = np.random.rand(50) * 100  # 50 valori casuali per l'asse y
sizes = np.random.randint(10, 100, size=50)  # Dimensioni dei punti

# Creare il grafico a dispersione
plt.scatter(x, y, s=sizes)

# Aggiungere etichette e titolo
plt.title('Grafico a Dispersione')
plt.xlabel('Valore X')
plt.ylabel('Valore Y')

# Mostrare la griglia
plt.grid()

# Mostrare il grafico
plt.show()
```

**istogramma**
mostra la distribuzione dei valori!


```python
import numpy as np
import matplotlib.pyplot as plt

# Generare dati casuali
dati = np.random.randn(1000)  # 1000 valori casuali con distribuzione normale

# Creare l'istogramma 
# bins - specifica il numero di intervalli (o "barre") nell'istogramma.
plt.hist(dati, edgecolor='black', alpha=0.7, bins=50) 

# Aggiungere etichette e titolo
plt.title('Istogramma')
plt.xlabel('Valore')
plt.ylabel('Frequenza')

# Mostrare la griglia
plt.grid(axis='y', alpha=0.75)

# Mostrare il grafico
plt.show()
```
