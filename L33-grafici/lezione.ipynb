```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
lista = np.random.randint(0,10, 30)
lista = lista + 100

plt.plot(lista)
plt.show()
```

```python
x = np.arange(10) + 2010
y = np.random.randint(0,10, 10)

plt.plot(x, y)
plt.show()
```

```python
step = 20
x = np.arange(step) + 2010
y = np.random.randint(0,10, step)
y2 = np.arange(0, step)

plt.title("grafo 1")
plt.xlabel("anni")
plt.ylabel("valori")

plt.plot(x, y, x, y2)
plt.show()
```

```python
valori = [15,30,45,10,25]
etichette = ["A", "B", "C", "D", "E"]
colori = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'] 

plt.title('Grafico a Torta con 5 Settori')
plt.pie(valori, labels=etichette, colors=colori, autopct='%1.1f%%')
plt.show()
```

```python
partiti = ["DC", "PSI", "PSD", "MXT", "CFZ"]
voti = [1000,1100,3450,2345,333]
colori = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'] 

for i,v in enumerate(voti):
    print(i,v)
    plt.text(i, v+30, str(v), ha='center')

plt.xlabel("partiti")
plt.ylabel("voti")
plt.title("votazioni")
plt.xticks(rotation=45)
plt.bar(partiti, voti, color=colori)
plt.tight_layout()
plt.show()
```

```python
x = np.random.rand(50) * 100
y = np.random.rand(50) * 200
plt.grid()
plt.scatter(x,y)
```

```python
x = np.random.rand(50) * 100
y = np.random.rand(50) * 200
sz = np.random.randint(10,100, size=50)

plt.grid()
plt.scatter(x,y, s=sz, alpha=0.5)
plt.show()
```

```python
dati = np.random.randn(1000)
plt.hist(dati, alpha=0.5, edgecolor='black', bins=50)
plt.grid(axis='y', alpha=0.5)
plt.show()
```

## pandas e grafici

```python
import pandas as pd
df = pd.read_csv('mercato_azionario.csv', sep=';')
df.head()
```

```python
df["Apertura"].plot()
```

- Grafico a Linee - kind='line'
- Grafico a Barre - kind='bar'
- Grafico a Barre Orizzontali - kind='barh'
- Istogramma - kind='hist'
- Grafico a Dispersione - kind='scatter'
- Grafico a Torta - kind='pie'
- Grafico a Area - kind='area'
- Box Plot - kind='box'
- Heatmap - (non utilizza kind, richiede seaborn)
- Pair Plot - (non utilizza kind, richiede seaborn)

```python
df["Apertura"].plot(kind='bar')
```

```python
df["Apertura"].rolling(3).mean()
```

```python
df["Apertura"].plot(kind='line')
df["Apertura"].rolling(3).mean().plot()

plt.savefig('grafo.jpg')

plt.show()
```
