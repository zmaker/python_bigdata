```python
from sklearn.datasets import make_regression

dataset = make_regression(n_samples = 500, n_features = 5, noise = 20)
```

```python
X = dataset[0]
y = dataset[1]

X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=100)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
previsione = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, previsione)
mse = mean_squared_error(y_test, previsione)
r2 = r2_score(y_test, previsione)

mae, mse, r2
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, previsione, color='blue', alpha=0.5)

plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
```
